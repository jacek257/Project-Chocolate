```python
import os
import pandas as pd
import fnmatch
```

```python
path='/home/ke/Desktop/SH'

# make sure the path ends with '/'
if path[-1] != '/':
    path += '/'
```

```python
# all grab all the .txt files in the endtidal folder
txt_files = [file for file in os.listdir(path) if file.endswith('.txt')]

#separate patient ID and scan date and pack into tuple
df = pd.DataFrame({
                'Cohort':[f[0:2] for f in txt_files],
                'ID':[f[3:6] for f in txt_files],
                'Date':[f[7:15] for f in txt_files],
                'Trailing': [f[15:] for f in txt_files]
             })
# df = df.set_index('ID')
df.sort_values(by='ID')

# df = pd.Series(txt_files)
# df = df.str.split(pat='_| ')
# df.columns = ['Cohort', 'ID', 'Date', 'Trailing']
# df
```

```python
b_path = '/home/ke/Desktop/SH_info'

# make sure the path ends with '/'
if b_path[-1] != '/':
    b_path += '/'
```

```python
for row in df.itertuples(index=True, name='Pandas'):
    patient = getattr(row, "Cohort")+getattr(row, "ID")
    bold_path = b_path+patient+'/BOLD/'
#     print(bold_path)    
    for root, dirs, files in os.walk(bold_path):
        for name in files:
            if fnmatch.fnmatch(name, '*a.nii'):
                bold_path += name
                print(bold_path)
                print(os.path.isfile(bold_path))
```
